### Step 5. Create a plot to check the relationship between horsepower and acceleration. Note down your insight for the same. Beautify the graph using various customizations.

```python

mpg_df.plot(kind = 'scatter', x = 'horsepower', y = 'acceleration')
plt.title('The Relationship between horsepower and acceleration',color='r',fontsize=15)
plt.ylabel('acceleration',color='green',fontsize=15)
plt.xlabel('horsepower',color='Orange',fontsize=15)
plt.show()
```

### Step 6. Generate subplots to display the histograms for acceleration, displacement and weight and kilometer_per_litre. Note down your insight for the same

```python
# Not found 'kilometer_per_litre' so used 'origin' column
```

```python

x = np.array(mpg_df['acceleration'])
y = np.array(mpg_df['displacement'])
z = np.array(mpg_df['weight'])

#plot 1:
plt.subplot(1, 3, 1)
plt.hist(x)

#plot 2:
plt.subplot(1, 3, 2)
plt.hist(y)

#plot 3:
plt.subplot(1, 3, 3)
plt.hist(z)
plt.suptitle('kilometer_per_litre',color='r',fontsize=20)

plt.show()
```
